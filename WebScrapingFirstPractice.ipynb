```python
from urllib.request import urlopen
import re

urls = ["http://google.com", "http://nytimes.com", "http://cnn.com"]
regex = '<title>(.+?)</title>'
pattern = re.compile(regex)

i=0
while(i<len(urls)):
	html = urlopen(urls[i])
	content = html.read().decode('utf-8')
	titles = re.findall(pattern, content)
	print(titles)
	i+=1
```

```python
import urllib.request import urlopen
import re

url = ""
regex = "<span id=''>(.+?)</span>"
pattern = re.compile(regex)

html = urlopen(url)
content = html.read()

text = re.findall(pattern, content)
print(text)
```

## Using BeautifulSoup Python Library

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
content = urlopen("http://nytimes.com")
soupobj = BeautifulSoup(content, "lxml")
print(soupobj.title.string)  #soupobj.title.name # .text (it can work if there is child tag as well)  # .get_text() 
```

```python
print(soupobj.findAll('p')) # .find_all() is also same
```

```python
allp = soupobj.findAll('p') # .find_all() is also same

## For Pringint all rows texts
'''
for para in allp:
    print(para.string + "\n")
'''
    
## For printing only first two rows

i=0
while i<2:
    print(allp[i].string)
    i+=1
```

```python
for link in soupobj.find_all("a"):
    print(link.get('href'))
```

```python
nav = soupobj.nav  #important
for a in nav.find_all("a"):
    print(a.text)
```

```python
for p in soupobj.find_all('h2', class_="visually-hidden"):
    print(p)
```

# scraping tables and XML (just note)

```python
table = soupobj.table #or soupobj.find('table')
tablerows = table.find_all("tr")

for tr in tablerows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```

```python
soup = BeautifulSoup(content, "xml")
for url in soup.find_all('loc'):
    print(url.text)
```
